```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
planets = sns.load_dataset('planets')
corr_matrix = planets.corr()
```

```python
labels = ['número', 'período\norbital', 'massa', 'distância', 'ano']
figsize = (9.6, 7.2)
```

```python
heatmap_1, ax = plt.subplots(figsize=figsize)

sns.heatmap(
    corr_matrix,
    ax=ax,
)

ax.set_xticklabels(labels)
ax.set_yticklabels(labels, verticalalignment='center')

heatmap_1.savefig('heatmap-1.png')
```

```python
# Neat use of sns.heatmap to generate colorbars.
cmaps = [
    ['coolwarm', 'PRGn', 'RdGy', 'PuOr'],
    ['viridis', 'plasma', 'Oranges', 'RdPu'],
]

colorbars, ax = plt.subplots(4, 2, figsize=(16, 6))

def create_colorbar(ax, colormap):
    sns.heatmap(
        np.array([np.linspace(-1, 1, 111)]),
        cmap=colormap,
        center=0,
        vmin=-1,
        vmax=1,
        cbar=False,
        xticklabels=[],
        yticklabels=[],
        ax=ax,
    )

for col_idx, cmap_list in enumerate(cmaps):
    for row_idx, colormap in enumerate(cmap_list):
        create_colorbar(ax[row_idx, col_idx], colormap)

colorbars.savefig('colormaps.png')
```

```python
heatmap_2, ax = plt.subplots(figsize=figsize)

sns.heatmap(
    corr_matrix,
    cmap='coolwarm',
    ax=ax,
)

ax.set_xticklabels(labels)
ax.set_yticklabels(labels, verticalalignment='center')

heatmap_2.savefig('heatmap-2.png')
```

```python
heatmap_3, ax = plt.subplots(figsize=figsize)

sns.heatmap(
    corr_matrix,
    cmap='coolwarm',
    center=0,
    vmin=-1,
    vmax=1,
    ax=ax
)

ax.set_xticklabels(labels)
ax.set_yticklabels(labels, verticalalignment='center')

heatmap_3.savefig('heatmap-3.png')
```

```python
cmap = sns.diverging_palette(
    h_neg=240,
    h_pos=10,
    s=100,
    # n=8,
    as_cmap=True,
)

heatmap_4, ax = plt.subplots(figsize=figsize)

sns.heatmap(
    corr_matrix,
    cmap=cmap,
    center=0,
    vmin=-1,
    vmax=1,
    square=True,
    linewidths=0.01,
    annot=True,
    ax=ax
)

ax.set_xticklabels(labels)
ax.set_yticklabels(labels, verticalalignment='center')

heatmap_4.savefig('heatmap-4.png')
```
